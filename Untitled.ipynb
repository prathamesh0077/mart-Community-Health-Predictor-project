```python
# ðŸŸ¢ Step 1: Import Libraries
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# ðŸŸ¢ Step 2: Load Your Dataset
df = pd.read_csv('datafile.csv')
```

```python

# OPTIONAL: Drop columns you don't want (like names or non-numeric)
df = df.drop(['District Names', 'State/UT'], axis=1)
```

```python
# ðŸŸ¢ Step 3: Handle Missing Values (you can also choose to drop)
df = df.fillna(df.mean(numeric_only=True))
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_df = scaler.fit_transform(df)
```

```python
# ðŸ§¹ Clean all string-formatted numbers (e.g., '(1261)') and convert to float
def clean_numeric(x):
    if isinstance(x, str):
        x = x.replace("(", "").replace(")", "").replace(",", "").strip()
    try:
        return float(x)
    except:
        return None

df = df.applymap(clean_numeric)

# ðŸ§¼ Fill any remaining missing values
df = df.fillna(df.mean(numeric_only=True))

# âœ… Now it's safe to scale
scaled_df = scaler.fit_transform(df)
```

```python


# ðŸŸ¢ Step 4: Normalize (Scale) the Features
scaler = MinMaxScaler()
scaled_df = scaler.fit_transform(df)
```

```python

# ðŸŸ¢ Step 5: Choose Number of Clusters using Elbow Method (Optional Visual)
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_df)
    wcss.append(kmeans.inertia_)


plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method to Choose Optimal Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python

# ðŸŸ¢ Step 6: Fit Final Model (choose 3-5 clusters based on elbow)
model = KMeans(n_clusters=4, random_state=42)
clusters = model.fit_predict(scaled_df)
```

```python
# âœ… Save feature names for later use in prediction
feature_names = df.columns.tolist()
```

```python
# Add cluster labels to the dataset (if not already done)
df['Cluster'] = model.labels_

# Compare average values for each cluster
cluster_summary = df.groupby('Cluster').mean(numeric_only=True)

# Display the cluster characteristics
import pandas as pd
pd.set_option("display.max_rows", None)  # Optional: show all rows
print(cluster_summary)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Reduce to 2D
pca = PCA(n_components=2)
components = pca.fit_transform(scaled_df)

# Plot clusters
plt.figure(figsize=(8, 6))
plt.scatter(components[:, 0], components[:, 1], c=model.labels_, cmap='viridis')
plt.title("Health Clusters Visualization")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.colorbar(label='Cluster')
plt.show()
```

```python

# ðŸŸ¢ Step 7: Save model components for user prediction
import joblib
joblib.dump(model, 'health_cluster_model.pkl')
joblib.dump(scaler, 'health_scaler.pkl')
df.to_csv('clustered_health_data.csv', index=False)
```

```python

# ðŸŸ¢ Step 8: Predict cluster from new user input
def predict_health_cluster(user_input_dict):
    # Fill missing features with default values
    for feature in feature_names:
        if feature not in user_input_dict:
            user_input_dict[feature] = 0  # or use df[feature].mean()

    # Create DataFrame with correct order
    user_df = pd.DataFrame([[user_input_dict[f] for f in feature_names]], columns=feature_names)

    # Scale input
    user_scaled = scaler.transform(user_df)

    # Predict cluster
    cluster = model.predict(user_scaled)[0]
    return cluster
```

```python
# ðŸŸ¢ Step 9: Example user input (you can customize as needed)

user_input = {
    'Sex ratio of the total population (females per 1,000 males)': 950,
    'All women age 15-19 years who are anaemic22 (%) ': 55.0,
    'All women age 15-49 years who are anaemic22 (%)': 60.0,
    'Average out-of-pocket expenditure per delivery in a public health facility (for last birth in the 5 years before the survey) (Rs.)': 800.0,
    'Births attended by skilled health personnel (in the 5 years before the survey)10 (%)': 85.0,
    # â—ï¸Continue adding all needed features or let the function handle missing ones
}

predicted_cluster = predict_health_cluster(user_input)
print("ðŸ§  Predicted Health Cluster:", predicted_cluster)
```
